```
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.tree import export_text
import numpy as np
import matplotlib.pyplot as plt
import graphviz

# [x_dif, y_dif, game_mode, label]
dataset = np.array([
    [1, -1, -1, 13],
    [0, -1, -1, 1],
    [-1, -1, -1, 14],

    [1, 0, -1, 3],
    [0, 0, -1, 0],
    [-1, 0, -1, 4],

    [1, 1, -1, 23],
    [0, 1, -1, 2],
    [-1, 1, -1, 24],

    [1, -1, 1, 24],
    [0, -1, 1, 2],
    [-1, -1, 1, 23],

    [1, 0, 1, 4],
    [0, 0, 1, 0],
    [-1, 0, 1, 3],

    [1, 1, 1, 14],
    [0, 1, 1, 1],
    [-1, 1, 1, 13]
])
print("Dataset = \n", dataset, "\n")

# [x_dif, y_dif, game_mode]
X = dataset[:, 0:3]
print("X = \n", X, "\n")

# [label]
y = dataset[:, 3]
print("y = \n", y, "\n")

decision_tree = DecisionTreeClassifier(criterion = 'entropy',
                                       random_state = 0,
                                       max_depth = 10)

decision_tree = decision_tree.fit(X, y)

r = export_text(decision_tree, feature_names = ["x_dif", "y_dif", "game_mode"])
print(r)

# [0:'hold', 1:'up', 2:'down', 3:'left', 4:'right', 13:'up-left', 14:'up-right', 23:'down-left', 24:'down-right']
tree_dot = export_graphviz(decision_tree,
                           feature_names = ['x_dif', 'y_dif', 'game_mode'],
                           class_names = ['hold', 'up', 'down', 'left', 'right', 'up-left', 'up-right', 'down-left', 'down-right'],
                           filled = True)
graph = graphviz.Source(tree_dot)
graph
```
